```python
from bs4 import BeautifulSoup
import os
```

```python
f = open("content_approved_doc.htm", "r")
contents = f.read()
f.close()

contents
```

```python
soup = BeautifulSoup(contents)
body = soup.findAll("body")
print(body[0].prettify())
```

```python
import re

def clean(text):
    #Remove styles
    text = re.sub(r"( style=\".+?\")", "", text)
    
    #Remove classes
    text = re.sub(r"( class=\".+?\")", "", text)
    
    #Remove span tags
    text = re.sub(r"</?span>", "", text)
    return text
```

```python
body_string = str(body[0])
body_string
```

```python
clean(body_string)
```
